```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
combined_df = pd.merge(mouse_metadata,study_results, how='outer', on='Mouse ID' )
# Display the data table for preview
combined_df.head()
```

```python
# Checking the number of mice.
mice_count = len(combined_df['Mouse ID'].unique())
mice_count
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicate_mice = combined_df[combined_df.duplicated(['Mouse ID','Timepoint'])]
duplicate_mice
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
duplicate_mice_df = combined_df[combined_df['Mouse ID'] == 'g989']
duplicate_mice_df
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
cleaned_mouse_study_df = combined_df[combined_df['Mouse ID'] != 'g989']
cleaned_mouse_study_df
```

```python
# Checking the number of mice in the clean DataFrame.
mice_count_cleaned = len(cleaned_mouse_study_df['Mouse ID'].unique())
mice_count_cleaned
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
mean_tumor_volume = cleaned_mouse_study_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].mean()
median_tumor_volume = cleaned_mouse_study_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].median()
variance_tumor_volume = cleaned_mouse_study_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].var()
standard_dv_tumor_volume = cleaned_mouse_study_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].std()
sem_tumor_volume = cleaned_mouse_study_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].sem()
# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary dataframe.

tumor_volume_df = pd.DataFrame({
    "Mean Tumor Volume":mean_tumor_volume,
    "Median Tumor Volume": median_tumor_volume,
    "Variance in Tumor Volume":variance_tumor_volume,
    "Standard Deviation in Tumor Volume":standard_dv_tumor_volume,
    "SEM": sem_tumor_volume
})

tumor_volume_df
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# Using the aggregation method, produce the same summary statistics in a single line
regimen_df = cleaned_mouse_study_df.groupby('Drug Regimen').agg({'Tumor Volume (mm3)':['mean','median','var','std','sem']})
regimen_df
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse

#Creating df for capomulin final tumor value for greatest timepoint
capomulin_df = cleaned_mouse_study_df[cleaned_mouse_study_df["Drug Regimen"] == "Capomulin"]
capomulin_max=capomulin_df.drop_duplicates("Mouse ID", keep="last")
#capomulin_max

##Creating df for ramicane final tumor value for greatest timepoint 
ramicane_df = cleaned_mouse_study_df[cleaned_mouse_study_df["Drug Regimen"] == "Ramicane"]
ramicane_max=ramicane_df.drop_duplicates("Mouse ID", keep="last")
#ramicane_max

#Creating df for infubinol final tumor value for greatest timepoint 
infubinol_df = cleaned_mouse_study_df[cleaned_mouse_study_df["Drug Regimen"] == "Infubinol"]
infubinol_max=infubinol_df.drop_duplicates("Mouse ID", keep="last")
#infubinol_max

#creating df for Ceftamin final tumor value for greatest timepoint
ceftamin_df = cleaned_mouse_study_df[cleaned_mouse_study_df["Drug Regimen"] == "Ceftamin"]
ceftamin_max=ceftamin_df.drop_duplicates("Mouse ID", keep="last")
#ceftamin_max

# Merge this group df with the original dataframe to get the tumor volume at the last timepoint
tumor_value = pd.concat([capomulin_max, ramicane_max,infubinol_max,
                        ceftamin_max]).sort_values(["Tumor Volume (mm3)"],ascending = True).reset_index()
tumor_value
```

```python
# Generate a box plot of the final tumor volume of each mouse across four regimens of interest
fig, ax1 = plt.subplots()
ax1.set_xticklabels(treatments)
ax1.set_ylabel('Final Tumor Volume (mm3)')
ax1.boxplot(tumor_volume,flierprops={'marker': 'o', 'markersize': 15, 'markerfacecolor': 'green'})
plt.show()
```

```python
# Generate a line plot of tumor volume vs. time point for a mouse treated with Capomulin

capomulin_mouse_tumor_vol = cleaned_mouse_study_df[cleaned_mouse_study_df["Mouse ID"] == "s185"]
x_axis = capomulin_mouse_tumor_vol["Timepoint"]
y_axis = capomulin_mouse_tumor_vol["Tumor Volume (mm3)"]

plt.plot(x_axis,y_axis)
plt.title("Mouse s185 Capomulin Treatment")
plt.xlabel("Timepoint")
plt.ylabel("Tumor Volume (mm3)")
plt.show()
```

```python
# Generate a scatter plot of average tumor volume vs. mouse weight for the Capomulin regimen
capomulin = cleaned_mouse_study_df[cleaned_mouse_study_df["Drug Regimen"] == "Capomulin"]

#gathering unique mice ID's for the Capomulin drug regimen
capomulin_unique_mice = capomulin["Mouse ID"].unique()

#creating empty lists to house mouse weight and average tumor volume
mouse_weight = []
average_tumor_volume = []

#for loop to gather weight and average tumor volume
for m in capomulin_unique_mice:
    avg_tumor_vol = cleaned_mouse_study_df.loc[cleaned_mouse_study_df["Mouse ID"] == m, "Tumor Volume (mm3)"].mean()
    average_tumor_volume.append(avg_tumor_vol)
    weight = cleaned_mouse_study_df.loc[cleaned_mouse_study_df["Mouse ID"] == m, "Weight (g)"].mean()
    mouse_weight.append(weight)

plt.scatter(mouse_weight,average_tumor_volume)
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")
plt.title("Average Tumor Volume vs. Mouse Weight for Capomulin Regimen")


plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen
correlation_coefficient = round(st.pearsonr(mouse_weight,average_tumor_volume)[0],2)
correlation_coefficient
```

```python
#plotting the same scatter plot as above
plt.scatter(mouse_weight,average_tumor_volume)
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")
plt.title("Average Tumor Volume vs. Mouse Weight for Capomulin Regimen")

#calculating the linear regression and line equation
(slope, intercept, rvalue, pvalue, stderr) = st.linregress(mouse_weight, average_tumor_volume)

regress_values = np.asarray(mouse_weight) * slope + intercept

line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))

#Adding the linear regression to the scatter plot
plt.plot(mouse_weight,regress_values,"r-")
plt.annotate(line_eq,(18,38),fontsize=15,color="red")
plt.show()
```
